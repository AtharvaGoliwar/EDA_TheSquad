```python
### The columns to be added are Pclass , Name, Ticket, Cabin
```

```python
new_df = new_df[['Pclass','Name','Ticket','Cabin']]
```

```python
df = pd.concat([df,new_df],axis=1)
```

```python
df = df.iloc[:,[8,0,2,3,4,5,9,10,7,1,6]]
```

```python
### Now we have our data properly set and have filtered the columns. Now its time to filter the rows
```

```python
#Checking for NA fields
df.isna().sum()
```

```python
#Dropped rows with NAN embarked fields
df = df.dropna(subset="Embarked")
```
